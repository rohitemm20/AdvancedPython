```python
odds = [1, 2, 3, 4]

# Unpacking the list
a, b, c, d = odds
print(a) 
print(b)  
print(c)  
print(d)  
```

```python

# a gets 1, b gets 2 and c gets the rest
a, b, *c = odds
print(a)  # 1
print(b)  # 3
print(c)  # [3, 4] 
```

```python
# Here we are  unpacking 3 elements: 
# 'a' gets the first item (1), 
# '*b' captures the middle items (2,3,5),
# and 'c' gets the last item (4)
a, *b, c = [1,2,3,5,4]
print(a)  
print(b)  
print(c)
```

```python
# A function that adds 3 no's
def xor(a, b, c):
    return (a + b + c)

num = [1, 5, 7]

# Unpacks the list and sends it as arguments: a(1,5,7)
print(xor(*num)) 
```

```python
# function that says hiii to each name in the list
def say_hi(*names):  # *names means it can take any number of names
    for n in names:
        print(f"hiii {n}")

say_hi("rohit", "sidhi", "manon")
```

```python
# A dictionary with two key-value pairs
person = {"name": "rohit", "age": 20}

# ['rohit', 20]
name, age = person.values()

print(name) 
print(age)   
```

```python
# Alternative way to create a dictionary
dic = {}
dic["name"] = "rohit"
dic["age"] = 20

print(dic)  

```
