```python
# a numeric list
num = [10, 60, 30, 40, 90]

# printing the elements in the list using a loop
for i in range(0, len(num)):
    print(f"Element at index {i} is {num[i]}")
```

```python
# string list
food = ["chocolate", "banana", "maple", "sugarcane", "litchi"]

# First 3 elements
print("First 2:", food[:3])

# Last 2 elements
print("Last 2:", food[-2:])

# From 2nd element to the last element
print("2nd to last element:", food[1:])
```

```python
# Simple list
moss = ["bluemoss","greenmoss","blackmoss"]
# Loop
for i in moss:
    print("moss is:", i)
```

```python
# reusing the moss list
moss = ["bluemoss","greenmoss","blackmoss", "purplemoss", "pinkmoss"]

# only looping through the middle 3 colors
for mos in moss[1:4]:
    print("Sliced Color:", mos)
```
