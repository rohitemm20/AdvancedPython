```python
import numpy as np


# 1. Creating a NumPy array using np.array()
a = np.array([10,20,30,40,50])
print(a)
```

```python
# 3. Creating a 3x3 Identity Matrix using np.eye()
imatrix = np.eye(3) 
print(imatrix)
```

```python
# 3. Creating a  evenly spaced numbers using np.linspace()
lspace = np.linspace(0, 20,5)
print(lspace)
```

```python
# 4. Creating a range of numbers using np.arange()
arange = np.arange(0, 24, 4) 
print(arange)
```

```python
# 5. Creating an array of zeros using np.zeros()
zero = np.zeros((4, 4))  # 4x4 matrix 
print(zero)
```

```python
# 6. Creating an array of ones using np.ones()
ones_arr = np.ones((6, 5))
print(ones_arr)
```

```python
# 7. Creating an array filled with a constant value using np.full()
six_arr = np.full((6, 6), 5) 
print(six_arr)
```

```python
# 8.producing Random numbers using np.random.rand()
rand_arr = np.random.rand(3, 3)
print(rand_arr)
```

```python
# 9. Random integers using np.random.randint()
rand_int = np.random.randint(1, 100, size=(3, 3))  
print(rand_int)
```

```python
# 10. Create indices grid using np.indices()
ind = np.indices((3, 3)) 
print(ind)
```

```python
# 11. Check array attributes (ndim, size, shape)
print("\nArray Attributes:")
print("ndim (number of dimensions):", a.ndim)
print("size (total number of elements):", a.size)
print("shape (rows, columns):", a.shape)
```

```python
# 12. Reshape an array using np.reshape()
reshaped = np.arange(12).reshape(6,2) 
print(reshaped)
```

```python
# 13. Transpose of a matrix using np.transpose()
transposed = reshaped.T
print(transposed)
```

```python
# 14. Mathematical operations
math_arr = np.array([10, 20, 30, 40, 50])
print("\nSum:", np.sum(math_arr))
print("Mean:", np.mean(math_arr))
print("Max:", np.max(math_arr))
print("Min:", np.min(math_arr))
print("Standard Deviation:", np.std(math_arr))
```

```python
# 15. Creating diagonal matrix using np.diag()
diag_arr = np.diag([10,10,10,10])
print(diag_arr)
```

```python
# 16. Flattening array using np.ravel()
flat = reshaped.ravel()
print(flat)
```

```python
# 17. Sorting array using np.sort()
unsorted = np.array([60, 100, 550, 20, 440])
sorted_arr = np.sort(unsorted)
print(sorted_arr)
```

```python
# 18. Unique values using np.unique()
dup_arr = np.array([1, 2, 2, 3, 4, 4, 5,6,4,3,5,9,6])
unique_vals = np.unique(dup_arr)
print(unique_vals)
```
