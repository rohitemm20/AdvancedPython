```python
# A PYTHON LOGIC THAT DOES NOT ALLOW MODIFYING THE LIST
class ReadOnlyList:
 def __init__(self, data):
 self._data = tuple(data) # Store as tuple for immutability
 def __getitem__(self, index):
 return self._data[index]
 def __len__(self):
 return len(self._data)
 def __iter__(self):
 return iter(self._data)
#EXAMPLE:
readonly = ReadOnlyList([1, 2, 3, 4])
print(readonly[0])
readonly[0] = 10 
```
